# TensorFlow 2 quickstart - Deployment

## Welcome to PrimeHub!

In this quickstart, we will perfome following actions to deploy the trained model: 
1. Deploy the trained model on PrimeHub <a target="_blank" href="https://docs.primehub.io/docs/model-deployment-feature">Model Deployments</a>.
1. Test deployed model.
1. (Advanced) Use PrimeHub <a target="_blank" href="https://github.com/InfuseAI/primehub-python-sdk">Python SDK</a> to deploy the trained model.

## 1. Deploy the trained model on PrimeHub <a target="_blank" href="https://docs.primehub.io/docs/model-deployment-feature">Model Deployments</a>

```python
SAVED_DIR = "~/phfs/example-models/tensorflow2"
LATEST_MODEL_NAME = ! ls -t $SAVED_DIR | head -1

if not isinstance(LATEST_MODEL_NAME, list) or len(LATEST_MODEL_NAME) == 0:
    print("There is no models yet. Please go to the previous quickstart example to get on.")
    raise
    
LATEST_MODEL_PATH = f"example-models/tensorflow2/{LATEST_MODEL_NAME[0]}"
print(LATEST_MODEL_PATH)
```

```python
print(f"Now, we've already prepared the ready-to-deploy model in {LATEST_MODEL_PATH}.")
```

Next, back to PrimeHub and select `Deployments`.

<img src="img/tensorflow2/3-1-menu.png"/>

In the `Deployments` page, click on `Create Deployment` button.

<img src="img/tensorflow2/3-2-create-deployment.png"/>

1. Fill in the [Deployment Name] field with [tf2-quickstart].
1. Select the [Model Image] field with [TensorFlow2 server]; this is a pre-packaged model server image that can serve TensorFlow 2 model.

<img src="img/tensorflow2/3-3-name-model-image.png"/>

```python
print(f"Fill in the [Model URI] field with [phfs:///{LATEST_MODEL_PATH}].")
```

<img src="img/tensorflow2/3-4-model-uri.png"/>

Choose the Instance Type, the minimal requirements in this quickstart is `CPU: 0.5 / Memory: 1 G / GPU: 0`.

<img src="img/tensorflow2/3-5-resource.png"/>

Then, click on `Deploy` button.

Our model is deploying, let's click on the `Manage` button of `tf2-quickstart` deployment cell.

In the deployment detail page, we can see the `Status` is `Deploying`.

<img src="img/tensorflow2/3-7-deploying.png"/>

Wait for a while and our model is `Deployed` now!
